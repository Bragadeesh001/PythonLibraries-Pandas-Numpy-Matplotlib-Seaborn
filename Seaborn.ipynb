# Seaborn tutorial

### Distribution plots
1. distplot
2. joinplot
3. pairplot

```python
# importing the libraries

import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head(25)
```

```python
df.dtypes
```

### Correlation with Heatmap
a correlation heatmap uses colored cells, typically in a monochromatic scale to a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
# in correlation value is gonna be ranging from -1 to +1

df.corr()
```

```python
#heat map

sns.heatmap(df.corr(),annot=True)
```

### Univariate analysis / Bivariet analysis


```python
# jointplot

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
# jointplot

sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pairplot
a 'pairs plot' is also known as scatterplot, in which one variable in the same data row is matched with another variables value, like this: pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
# pair plot
# feature has to int or float

sns.pairplot(df)
```

```python
#pair plot with categorical feature

sns.pairplot(df,hue='sex')
```

### Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.displot(df['tip'])
```

## Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points

1. boxplot
2. violinplot
3. countplot
4. barplot

```python
## count plot

sns.countplot('sex',data=df)
```

```python
## count plot
## bivarier

sns.countplot('sex',data=df,hue='smoker')
```

```python
## bar plot

sns.barplot(x='sex',y='total_bill',data=df)
```

### Box plot
A box and whisker plot(sometimes called a boxplot) is a graph that present information from a five number summary

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,hue='smoker')
```

### Violin plot
violin plot helps us to see both the distrubution of data in terms of kernel density estimation and the box plot


```python
sns.violinplot(x='total_bill',y='day',data=df)
```
