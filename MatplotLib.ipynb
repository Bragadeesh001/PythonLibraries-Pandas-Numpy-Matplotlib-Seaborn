# MatplotLib
MatplotLib is a plotting library for the Python programming language and its numerical mathematics extension Numpy. It provides an object-oriented API for embedding plots into applications using general-purpose Gui toolkits like Tkinter, wxPython, Qt, or GTK+


Some of the major Pros
1. Generally easy to get started for simple plots
2. Support for custom labels and texts
3. Great control of every element in a figure
4. High-quality output in many formats
5. Very customizable in General

```python
#importing libraries
import matplotlib.pyplot as plt

#if we dont want to use plt.show then we use this
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
## plotting using matplotlib

#plotting scatter plot

plt.scatter(x,y,c='g')
plt.title('Scatter Plot')
plt.xlabel('x Axis')
plt.ylabel('y Axis')

# to save this image
plt.savefig('scatter plot')
```

```python
##plt plot

y=x**2
plt.plot(x,y,'r*--')
```

```python
# Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.subplot(2,2,4)
plt.plot(x,y,'y*--')
```

```python
# for plotting sine graphs

x=np.arange(0,4 * np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y_sin)
plt.subplot(1,2,2)
plt.plot(x,y_cos)
```

```python
# bar plot

x=[1,2,5]
y=[5,6,9]

plt.bar(x,y)
```

```python
## Histograms
# y axis is the density i.e the frequency of that value repeted
# x axis is the bins i.e how many numbers present between

x=[1,5,7,8,11,6,8,8,5,4,3]
plt.hist(x)
```

```python
# Box plot

data=[1,2,4,5,6,7,8,2,3,5,20]

plt.boxplot(data)
```

```python
# pie plot

plt.pie(x=[50,30,10],labels=['a','b','c'],autopct='%1.2f%%',shadow=True,explode=[0.02,0.1,0.1],startangle=90)
```
