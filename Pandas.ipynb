## csv

```python
data=('col1,col2,col3\n'
         'x,y,1\n'
         'a,b,2\n'
         'c,d,3\n')
```

```python
type(data)
```

```python
from io import StringIO
```

```python
StringIO()
```

```python
pd.read_csv(StringIO(data))
```

```python
# if i want only specific column when reading the csv file
# usecols is used

df=pd.read_csv(StringIO(data),usecols=['col1','col2'])
df
```

```python
# for saving to csv file
# to_csv('filename')
#df.to_csv('test.csv')
```

```python
# for providing custom datatype to a csv file then

df=pd.read_csv(StringIO(data),dtype=object)
df
```

```python
df.info()
```

```python
## we can specify individual column with different datatype

df=pd.read_csv(StringIO(data),dtype={'col1':str,'col2':object,'col3':int})
df
```

```python
df.dtypes
```

```python
# to make the first column as index

df=pd.read_csv(StringIO(data),index_col=0)
df
```

```python
# to make the second column as index

df=pd.read_csv(StringIO(data),index_col=1)
df
```

```python
## Quoting and Escape Characters. Very useful in NLP

data='a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```
